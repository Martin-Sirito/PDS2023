```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
# funcion Seno
def mi_funcion_sen( vmax, dc, ff, ph, nn, fs):
    
    Ts = 1/fs
    tt = np.arange(0, nn*Ts, Ts)
    ww = 2 * np.pi * ff
    
    xx = vmax * np.sin(ww * tt + ph) + dc
    
    return tt, xx
```

Probamos la funcion

```python
tt, xx = mi_funcion_sen(vmax = 1, dc = 0, ff = 1, ph = 0, nn = 1000, fs = 1000)
plt.plot(tt, xx)
```

```python
tt, xx = mi_funcion_sen(vmax = 380, dc = 0, ff = 2, ph = 0, nn = 1000, fs = 1000)
plt.plot(tt, xx)

tt, xx = mi_funcion_sen(vmax = 380, dc = 0, ff = 2, ph = 2/3 * np.pi, nn = 1000, fs = 1000)
plt.plot(tt, xx)

tt, xx = mi_funcion_sen(vmax = 380, dc = 0, ff = 2, ph = 4/3 * np.pi, nn = 1000, fs = 1000)
plt.plot(tt, xx)
```

```python
tt, xx = mi_funcion_sen(vmax = 1, dc = 0, ff = 1, ph = 0, nn = 8, fs = 7)
plt.plot(tt, xx, marker='o', linestyle=':', linewidth=2, markersize=5)
```

```python
def mi_funcion_cuad( vmax, dc, ff, ph, nn, fs, ton=0.5):
    
    #Tiempo
    Ts = 1/fs                       #Periodo de muestreo
    T_expl = Ts*nn                  #Tiempo de exploracion
    ww = 2 * np.pi * ff             #Frecuencia angular
    tt = np.arange(0, T_expl, Ts)   #base de tiempo
    
    xx = vmax * signal.square(ww * tt + ph, duty=ton) + dc
    
    return tt, xx
```

```python
tt, xx = mi_funcion_cuad(vmax = 10, dc = -5, ff = 5, ph = 1, nn = 1000, fs = 1000, ton=0.2)
plt.plot(tt, xx)
```
